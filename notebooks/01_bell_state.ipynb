# EXPERIMENT 01: Bell State Generation and Analysis

## Objective

The objective of this experiment is to generate a Bell State using Qiskit and analyze its measurement outcomes under ideal simulation conditions.

This experiment serves as the baseline for future quantum noise and error mitigation studies.

## Theoretical Background

A Bell State is a maximally entangled two-qubit quantum state.

Starting from the initial state |00⟩:

1. Apply a Hadamard gate to qubit 0.
2. Apply a CNOT gate with qubit 0 as control and qubit 1 as target.

This produces:

$$\frac{|00\rangle + |11\rangle}{\sqrt{2}}$$

The resulting state exhibits quantum entanglement, meaning the measurement outcome of one qubit is correlated with the outcome of the other.

```python
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()
qc.draw("mpl")
plt.savefig("../results/bell_circuit.png")
```

## Expected Results

The Bell State contains only two basis states:

- |00⟩
- |11⟩

Therefore:

P(00) = 50%
P(11) = 50%

The states |01⟩ and |10⟩ should not appear in an ideal simulation.

## Ideal Simulation

```python
from qiskit_aer import AerSimulator
simulator = AerSimulator()
result = simulator.run(qc, shots = 1000).result()
counts = result.get_counts()
print(counts)
```

### Measuremrnt results:

```python
print("Measurement Counts:")
print(counts)
```

### Visualization:

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
plt.savefig("../results/bell_histogram.png")
```

## Analysis

The histogram shows that only the states 00 and 11 are measured.

The approximately equal distribution confirms that the Bell State exists as an equal superposition of these two basis states.

No occurrences of 01 or 10 were observed, indicating correct circuit implementation under ideal conditions.

## Significance to Quantum Noise Analysis

The Bell State is highly sensitive to noise and decoherence.

In future experiments, various noise channels such as:

- Bit-flip noise
- Phase-flip noise
- Depolarizing noise

will be introduced to study how quantum information degrades.

The results obtained in this experiment serve as the ideal reference against which noisy and mitigated circuits will be compared.

## Conclusion

A Bell State was successfully generated and simulated using Qiskit.

The measurement results matched theoretical expectations, producing only the outcomes 00 and 11 with approximately equal probabilities.

This experiment establishes the baseline for subsequent quantum noise and error mitigation studies.
